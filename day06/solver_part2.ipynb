```python
import copy
```

```python
map = []

with open('input.txt', 'r') as file:
    for line in file:
        map.append(line.strip())
```

```python
# Pour toutes les lignes, split caractère par caractère
for i in range(len(map)):
    map[i] = list(map[i])
```

```python
def move_gardien(map, pos, direction):
    new_map = copy.deepcopy(map)
    new_pos = [pos[0], pos[1]]
    if direction == '^':
        new_pos[0] -= 1
    elif direction == 'v':
        new_pos[0] += 1
    elif direction == '<':
        new_pos[1] -= 1
    elif direction == '>':
        new_pos[1] += 1
    new_map[pos[0]][pos[1]] = 'X'
    # On vérifie que le gardien ne sort pas de la map
    if new_pos[0] < 0 or new_pos[0] >= len(map) or new_pos[1] < 0 or new_pos[1] >= len(map[0]):
        return False, new_map, new_pos
    
    new_map[new_pos[0]][new_pos[1]] = direction
    return True, new_map, new_pos
```

```python
def gardien_is_on_map(map, gardien):
    return gardien[0] >= 0 and gardien[0] < len(map) and gardien[1] >= 0 and gardien[1] < len(map[0])

def get_gardien_direction(map, gardien):
    if not gardien_is_on_map(map, gardien):
        raise ValueError(f"Position du gardien invalide : {gardien}")
    return map[gardien[0]][gardien[1]]

def get_gardien_next_position(gardien, direction, check_bounds=False, map_size=None):
    new_pos = [gardien[0], gardien[1]]
    if direction == '^':
        new_pos[0] -= 1
    elif direction == 'v':
        new_pos[0] += 1
    elif direction == '<':
        new_pos[1] -= 1
    elif direction == '>':
        new_pos[1] += 1

    if check_bounds and map_size:
        if new_pos[0] < 0 or new_pos[0] >= map_size[0] or new_pos[1] < 0 or new_pos[1] >= map_size[1]:
            return None  # Position invalide

    return new_pos

def change_gardien_direction(map, gardien):
    new_map = copy.deepcopy(map)
    if map[gardien[0]][gardien[1]] == '^':
        new_map[gardien[0]][gardien[1]] = '>'
    elif map[gardien[0]][gardien[1]] == 'v':
        new_map[gardien[0]][gardien[1]] = '<'
    elif map[gardien[0]][gardien[1]] == '<':
        new_map[gardien[0]][gardien[1]] = '^'
    elif map[gardien[0]][gardien[1]] == '>':
        new_map[gardien[0]][gardien[1]] = 'v'
    return new_map

def move_forward(map, gardien):
    # Récupérer la direction et la prochaine position
    direction = get_gardien_direction(map, gardien)
    new_gardien = get_gardien_next_position(gardien, direction)
    
    # Si le gardien sort de la carte
    if not gardien_is_on_map(map, new_gardien):
        map[gardien[0]][gardien[1]] = 'X'  # Marque la position précédente
        return map, gardien  # Retourne sans déplacement
    
    # Déplacer le gardien
    map[gardien[0]][gardien[1]] = 'X'  # Marque l'ancienne position
    map[new_gardien[0]][new_gardien[1]] = direction  # Place le gardien
    return map, new_gardien
```

```python
# Stratégie 2 # Find gardien position
for row in range(len(map)):
    for col in range(len(map[row])):
        if map[row][col] in ['^', 'v', '<', '>']:
            gardien = (row, col)
            break

INITIAL_GARDIEN = copy.deepcopy(gardien)
INITIAL_MAP = copy.deepcopy(map)

infinite_loop = 0

for row in range(len(map)):
    for col in range(len(map[row])):
        print(f"row: {row} col: {col}")

        # Reset map
        map = copy.deepcopy(INITIAL_MAP)
        gardien = copy.deepcopy(INITIAL_GARDIEN)

        last_positions = set()
        last_positions.add((gardien[0], gardien[1], get_gardien_direction(map, gardien)))

        # Si la case est vide, placer un obstacle
        if map[row][col] == '.':
            map[row][col] = '#'

            while True:
                direction = get_gardien_direction(map, gardien)
                next_row, next_col = get_gardien_next_position(gardien, direction)

                if next_row < 0 or next_row >= len(map) or next_col < 0 or next_col >= len(map[0]):
                    break

                if not gardien_is_on_map(map, (next_row, next_col)):
                    break

                # Check si on a déjà été à cette position avec cette direction
                if (gardien[0], gardien[1], direction) in last_positions:
                    infinite_loop += 1
                    print(infinite_loop)
                    break

                if map[next_row][next_col] == '#':
                    map = change_gardien_direction(map, gardien)
                else:
                    map, gardien = move_forward(map, gardien)

                # Ajouter la position actuelle dans la liste des positions
                last_positions.add((gardien[0], gardien[1], direction))


```
