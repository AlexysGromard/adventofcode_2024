```python
import copy
```

```python
map = []

with open('input.txt', 'r') as file:
    for line in file:
        map.append(line.strip())
```

```python
# Pour toutes les lignes, split caractère par caractère
for i in range(len(map)):
    map[i] = list(map[i])
```

```python
for row in map:
    print(''.join(row))
```

```python
def move_gardien(map, pos, direction):
    new_map = copy.deepcopy(map)
    new_pos = [pos[0], pos[1]]
    if direction == '^':
        new_pos[0] -= 1
    elif direction == 'v':
        new_pos[0] += 1
    elif direction == '<':
        new_pos[1] -= 1
    elif direction == '>':
        new_pos[1] += 1
    new_map[pos[0]][pos[1]] = 'X'
    # On vérifie que le gardien ne sort pas de la map
    if new_pos[0] < 0 or new_pos[0] >= len(map) or new_pos[1] < 0 or new_pos[1] >= len(map[0]):
        return False, new_map, new_pos
    
    new_map[new_pos[0]][new_pos[1]] = direction
    return True, new_map, new_pos
```

```python
# Find gardien position
for row in range(len(map)):
    for col in range(len(map[row])):
        if map[row][col] in ['^', 'v', '<', '>']:
            gardien = (row, col)
            break

INITIAL_GARDIEN = copy.deepcopy(gardien)
INITIAL_MAP = copy.deepcopy(map)

#Variable boucle infinie
infiniteloop = 0

#  On va tester de rajouter un obstacle a toutes les positions et de tenter une boucle infinie
for row in range(len(map)):
    for col in range(len(map[row])):
        # Reset map
        map = copy.deepcopy(INITIAL_MAP)
        gardien = copy.deepcopy(INITIAL_GARDIEN)
        list_of_moves = [gardien, map[gardien[0]][gardien[1]]]

        print(f"row: {row} col: {col} et Infiniteloop: {infiniteloop}")

        # Si la case est vide
        if map[row][col] == '.':

            # Placer l'obstacle
            map[row][col] = '#'

            move_count = 0

            # Tant que le gardien est sur la map
            while gardien[0] >= 0 and gardien[0] < len(map) and gardien[1] >= 0 and gardien[1] < len(map[0]):
                # Check si le gardien est dans une boucle infinie
                if [gardien, map[gardien[0]][gardien[1]]] in list_of_moves:
                    # On a trouvé une boucle
                    infiniteloop += 1
                    break

                # Si le gardien est vers le haut
                if map[gardien[0]][gardien[1]] == '^':
                    # Si voisin du haut est un obstacle
                    if gardien[0] - 1 >= 0 and map[gardien[0] - 1][gardien[1]] == '#':
                        # Tourner de 90° à droite
                        map[gardien[0]][gardien[1]] = '>'
                    else:
                        # Avancer
                        res, map, gardien = move_gardien(map, gardien, '^')
                        if res == False:
                            break
                # Si le gardien est vers le bas
                elif map[gardien[0]][gardien[1]] == 'v':
                    # Si voisin du bas est un obstacle
                    if gardien[0] + 1 < len(map) and map[gardien[0] + 1][gardien[1]] == '#':
                        # Tourner de 90° à droite
                        map[gardien[0]][gardien[1]] = '<'
                    else:
                        # Avancer
                        res, map, gardien = move_gardien(map, gardien, 'v')
                        if res == False:
                            break

                # Si le gardien est vers la gauche
                elif map[gardien[0]][gardien[1]] == '<':
                    # Si voisin de gauche est un obstacle
                    if gardien[1] - 1 >= 0 and map[gardien[0]][gardien[1] - 1] == '#':
                        # Tourner de 90° à droite
                        map[gardien[0]][gardien[1]] = '^'
                    else:
                        # Avancer
                        res, map, gardien = move_gardien(map, gardien, '<')
                        if res == False:
                            break
                # Si le gardien est vers la droite
                elif map[gardien[0]][gardien[1]] == '>':
                    # Si voisin de droite est un obstacle
                    if gardien[1] + 1 < len(map[0]) and map[gardien[0]][gardien[1] + 1] == '#':
                        # Tourner de 90° à droite
                        map[gardien[0]][gardien[1]] = 'v'
                    else:
                        # Avancer
                        res, map, gardien = move_gardien(map, gardien, '>')
                        if res == False:
                            break

                print(gardien)
```
